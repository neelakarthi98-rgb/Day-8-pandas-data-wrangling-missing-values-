/////Employee Performance Analysis////

```python
import  pandas as pd
employee_details={'employee_id':['101','102','103','104'],
                    'name':['Alice','bob','charlie','MARK'],
                    'department':['HR','IT','HR','IT']}
performance_details={'employee_id':['101','102','103','104'],
                     'performance':[4.5,4,5,3.5]}
df1=pd.DataFrame(employee_details)
df2=pd.DataFrame(performance_details)
print(df1)
print(df2)

```

////calculate the average performance score per department////

```python
merge=pd.merge(df1,df2,how='inner',on=None)
print(merge)
avg=merge.groupby('department')['performance'].mean()
print(avg)
```
