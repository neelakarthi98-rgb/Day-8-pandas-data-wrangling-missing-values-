/////combining and cleaning sales data//////

```python
import pandas as pd
import numpy as np
sales_region_A={'PRODUCT_ID':['101','102','103','104'],
                'sales':[100,150,np.nan,300],
                'region':['A','A','A','A']}
sales_region_B={'PRODUCT_code':['101','102','103','104'],
                'Revenue':[100,np.nan,230,500],
                'region':['B','B','B','B']}
df1=pd.DataFrame(sales_region_A)
df2=pd.DataFrame(sales_region_B)
print(df1)
print()
print(df2)
                


```

//////standardize and merge/////

```python
df2.rename(columns={'PRODUCT_code':'PRODUCT_ID'},inplace=True)
print(df2)
merge=pd.concat([df1,df2],ignore_index=True)
print(merge)
```

/////clean and combined data//////

```python
miss_value=merge.isnull().sum()
print(miss_value)
print()
print("fill the missing values")
df1['sales']=df1['sales'].fillna(df1['sales'].mean())
print(df1)
print()
print("Drop the missing values")
print(df2)
drop=df2.dropna(subset=['Revenue'])
print(drop)
```

//////sales Analyze////

```python
df2.rename(columns={'Revenue':'sales'},inplace=True)
print("After changing region B revenue into sales")
print(df2)
print("fill the missing values in region B")
df2['sales']=df2['sales'].fillna(df2['sales'].mean())
print(df2)
merge_sales=pd.concat([df1,df2],ignore_index=True)
print("mergeing both columns")
print(merge_sales)
print("using groupby function to get total sales")
total_sales=merge_sales.groupby('PRODUCT_ID')['sales'].sum()
print(total_sales)
```
