/////Customer segmentation data prep////

```python
import pandas as pd
import numpy as np
customer_data={'customer_type':['new','regular',np.nan,'new','regular'],
'age':[25,31,20,np.nan,35]}
df=pd.DataFrame(customer_data)
print(df)
```

/////Impute Numerical Data////

```python
df['age']=df['age'].fillna(df['age'].median())
print(df)
```

////Handle categorical Data////

```python
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
print("categorical column",categorical_columns.tolist())
print()
df[categorical_columns]=df[categorical_columns].fillna('unknown')
print(df)

```

///one-hot encoding////

```python
df.encode=pd.get_dummies(df,columns=['customer_type'])
print(df.encode)
```
