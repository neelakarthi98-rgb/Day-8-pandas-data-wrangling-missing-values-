/////movie rating Data prep///

```python
import pandas as pd
import numpy as np
movie_rating={"movie_name":['inception','Titanic','the matrix'],
              'ratings':['4.4',np.nan,'3.5']}
movie_info={"movie_name":['inception','Titanic','the matrix'],
            "release_year":['1997','2000','2001']}
df1=pd.DataFrame(movie_rating)
df2=pd.DataFrame(movie_info)
print(df1)
print()
print(df2)
```

///fill na missing values///

```python
df1['ratings']=df1['ratings'].fillna(0)
print(df1)
```

////standardize string entries///

```python
df1['movie_name']=df1['movie_name'].str.lower().str.strip()
print(df1)
print()
df2['movie_name']=df2['movie_name'].str.lower().str.strip()
print(df2)
print("\nmerging two sets")
print()
merge=pd.merge(df1,df2,how='inner',on=None)
print(merge)
```
